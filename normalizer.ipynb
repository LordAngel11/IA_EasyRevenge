```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv("data/muestrafinal.csv")
dataset.head()
```

```python
# Verificar duplicados en la columna "latitud"
duplicados_latitud = dataset['Latitud'].duplicated()
print("Duplicados en Latitud:")
print(dataset[duplicados_latitud])

# Verificar duplicados en la columna "longitud"
duplicados_longitud = dataset['Longitud'].duplicated()
print("Duplicados en longitud:")
print(dataset[duplicados_longitud])
```

```python
# Eliminar los renglones con valores vacíos en la columna "latitud"
dataset = dataset.dropna(subset=['Latitud','Longitud','bedrooms','bathrooms','area'])
print(dataset)

```

```python
def augment_geographical(df):
    with open('../Datosgeo/school.geojson','r') as f:
        database = json.load(f)

    feature_types = {}
    for feature in database['features']:
        amenity_type = feature['properties'].get('amenity','na')
        if amenity_type in feature_types:
            feature_types[amenity_type] += 1
        else:
            feature_types[amenity_type] = 1

    cols = [name for name,val in sorted(list(feature_types.items()), key=lambda x : x[1])[::-1] if val > 100]
    points_np = df[['Latitud (Decimal)', 'Longitud (Decimal)']].to_numpy()

    item_coords = []
    item_tags = []
    for feature in database['features']:
        amenity_type = feature['properties'].get('amenity','na')
        if amenity_type not in cols: continue

        if feature['geometry']['type'] == 'Polygon':
            polygons = feature['geometry']['coordinates']
            polygon_centers = []
            for polygon in polygons:
                coords = np.array(polygon)
                polygon_centers.append(np.mean(coords.T,axis=1)[::-1])

            polygon_centers = np.array(polygon_centers)
            point_coords = np.mean(polygon_centers.T,axis=1)

            item_tags.append(amenity_type)
            item_coords.append(point_coords)

        elif feature['geometry']['type'] == 'Point':
            coords = feature['geometry']['coordinates']
            point_coords = np.array(coords)[::-1]

            item_tags.append(amenity_type)
            item_coords.append(point_coords)

    item_coords = np.array(item_coords)

    CYTHON_DISTANCE = lambda x,y : vincenty(x[0],x[1],y[0],y[1])
    btree = BallTree(item_coords,metric=CYTHON_DISTANCE)

    histograms = []
    for i,point in enumerate(points_np):
        print(f'Working in point ({point[0]},{point[1]}) {i+1}/{points_np.shape[0]}',end='     \r')
        found_points = []
        histogram = {col_name:0 for col_name in cols}

        try:
            found_points = btree.query_radius([point], r=1.5)[0]
        except ValueError:
            histograms.append(histogram)
            continue

        for p in found_points:
            if item_tags[p] in histogram:
                histogram[item_tags[p]] += 1
        histograms.append(histogram)

    geo_data = pd.DataFrame(histograms)

    return pd.concat([df,geo_data.set_index(df.index)], axis=1)
    #return pd.concat([df,geo_data], axis=1)
    #return geo_data

def add_distance_to_coast(df):
    augmented_df = df.copy()

    with open('../data/json/sealine.geojson', 'r') as f:
        database = json.load(f)

    item_coords = []
    for feature in database['features']:
        if feature['geometry']['type'] == 'LineString':
            line_coords = feature['geometry']['coordinates']
            for i in line_coords:
                if not any(np.isnan(i)):
                    item_coords.append([i[1],i[0]])

    item_coords = np.array(item_coords)
    CYTHON_DISTANCE = lambda x,y : vincenty(x[0],x[1],y[0],y[1])
    tree = BallTree(item_coords, metric=CYTHON_DISTANCE)

    augmented_df = augmented_df.dropna(subset=['Latitud (Decimal)', 'Longitud (Decimal)'])
    points_np = augmented_df[['Latitud (Decimal)', 'Longitud (Decimal)']].values

    dist, points = tree.query(points_np, k=1)
    augmented_df['distance_to_coast km'] = dist

    return augmented_df
```
